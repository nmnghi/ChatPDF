```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu từ các file CSV
file_paths = {
    "Euclidean Distance top 1": "../experimental/top_1_Euclidean Distance.csv",
    "Euclidean Distance top 3": "../experimental/top_3_Euclidean Distance.csv",
    "Euclidean Distance top 5": "../experimental/top_5_Euclidean Distance.csv",
}

dataframes = {key: pd.read_csv(path) for key, path in file_paths.items()}

# Vẽ biểu đồ
plt.figure(figsize=(16, 8))

# Lấy danh sách threshold
thresholds = dataframes["Euclidean Distance top 1"]["Threshold"]

t1, t2, t3 = (
    dataframes["Euclidean Distance top 1"],
    dataframes["Euclidean Distance top 3"],
    dataframes["Euclidean Distance top 5"],
)

f1 = {
    "Euclidean Distance top 1": t1["Precision"],
    "Euclidean Distance top 3": t2["Precision"],
    "Euclidean Distance top 5": t3["Precision"],
}

# Vẽ biểu đồ đường
plt.plot(thresholds, f1["Euclidean Distance top 1"], linewidth=2, label="Top 1")
plt.plot(thresholds, f1["Euclidean Distance top 3"], linewidth=2, label="Top 3")
plt.plot(thresholds, f1["Euclidean Distance top 5"], linewidth=2, label="Top 5")

# Gắn nhãn
plt.xlabel("Threshold")
plt.ylabel("Precision")
plt.title("Precision Across Different Top_k of Euclidean Distance")

# plt.xlim(0.69, 1.01)
plt.xlim(0.71, 0.19)
plt.xticks(thresholds, rotation=45)

plt.yticks(np.arange(0.0, 1.1, 0.1))

plt.legend()
plt.grid(True, linestyle="--", alpha=0.7)

# Điều chỉnh layout để tránh bị cắt nhãn
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
