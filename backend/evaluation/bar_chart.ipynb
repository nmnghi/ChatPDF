```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu từ các file CSV
file_paths = {
    "Cosine similarity": "../experimental/top_3_Cosine similarity.csv",
    # "Euclidean Distance": "../experimental/top_3_Euclidean Distance.csv",
    "Pearson Corr": "../experimental/top_3_Pearson Corr.csv",
}

dataframes = {key: pd.read_csv(path) for key, path in file_paths.items()}

# Vẽ biểu đồ
plt.figure(figsize=(16, 8))

# Lấy danh sách threshold
thresholds = dataframes["Cosine similarity"]["Threshold"]

t1, t2 = (
    dataframes["Cosine similarity"],
    # dataframes["Euclidean Distance"],
    dataframes["Pearson Corr"],
)

f1 = {
    "Cosine similarity": t1["F1-Score"],
    # "Euclidean Distance": t2["F1-Score"],
    "Pearson Corr": t2["F1-Score"],
}

# Xác định vị trí cột trên trục x
x = np.arange(len(thresholds))
width = 0.3

# Vẽ biểu đồ cột
plt.bar(x, f1["Cosine similarity"], width, label="Cosine similarity")
# plt.bar(x, f1["Euclidean Distance"], width, label="Euclidean Distance")
plt.bar(x + width, f1["Pearson Corr"], width, label="Pearson Corr")

# Gắn nhãn
plt.xlabel("Threshold")
plt.ylabel("F1-Score")
plt.title("F1-Score Between Cosine similarity and Pearson Corr in Top 3")
plt.xticks(x, thresholds, rotation=45)
plt.yticks(np.arange(0.0, 1.1, 0.1))
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Điều chỉnh layout để tránh bị cắt nhãn
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Đọc dữ liệu từ các file CSV
df = pd.read_csv("../experimental/top_3_Cosine similarity.csv")

# Vẽ biểu đồ
plt.figure(figsize=(16, 8))

# Lấy danh sách threshold
thresholds = df["Threshold"]

# Xác định vị trí cột trên trục x
x = np.arange(len(thresholds))
width = 0.3

# Vẽ biểu đồ cột
plt.bar(x, df["F1-Score"], width, label="Cosine similarity")

# Gắn nhãn
plt.xlabel("Threshold")
plt.ylabel("F1-Score")
plt.title("F1-Score of Cosine similarity in Top 3")
plt.xticks(x, thresholds, rotation=45)
plt.yticks(np.arange(0.0, 1.1, 0.1))
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Điều chỉnh layout để tránh bị cắt nhãn
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
